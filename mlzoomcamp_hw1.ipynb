```python
rolls = cars[cars.Make == 'Rolls-Royce']
```

```python
r2 = rolls[["Engine HP", "Engine Cylinders", "highway MPG"]]
```

```python
r2.drop_duplicates()
```

```python
a2 = r2.drop_duplicates().to_numpy()
```

```python
a2
```

```python
x = a2
```

```python
xtx = x.T.dot(x)
```

```python
xtx
```

```python
xtx_inv = np.linalg.inv(xtx)
xtx_inv
```

```python
xtx_inv.sum()
```
